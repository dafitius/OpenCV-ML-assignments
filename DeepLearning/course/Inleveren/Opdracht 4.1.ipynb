```python
# Import libraries
# numpy for array and matrices
# matplotlib visualization

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# sklearn for data sets

from sklearn.datasets import make_circles
```

```python
# returns array of coordinates to make a circle
X, y = make_circles(n_samples=1000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
```

```python
# Print x coordinates
X
```

```python
# returns amount of indicis with amount of elements at each index
X.shape
```

```python

#Makes a figure of 5 by 5
plt.figure(figsize=(5, 5))
# Plots a symbol at coordinates. 'ob' means o circle and b mean blue. 'xr' means x cross and r red
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
#limit of axis
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
#Makes a legend for the plotted symbols
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
# groups a liniear stack of layers on a model
model = Sequential()
```

```python
# Adds a densely-connected NN layer to the model
model.add(Dense(4, input_shape=(2,), activation='tanh'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
# configures the model for learning
model.compile(SGD(learning_rate=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
#trains the model
model.fit(X, y, epochs=20)
```

```python

# Makes a equally liniear interval between two points with a nummber of lines
hticks = np.linspace(-1.5, 1.5, 101)
vticks = np.linspace(-1.5, 1.5, 101)
# Makes a coordinates matrix with a horizontal and vertical interval
aa, bb = np.meshgrid(hticks, vticks)
# Translates slice objects to concatenation along the second axis.
ab = np.c_[aa.ravel(), bb.ravel()]
# Generates output predictions for the input samples.
c = model.predict(ab)
# Reshapes a tensor
cc = c.reshape(aa.shape)
```

```python
plt.figure(figsize=(5, 5))
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```
