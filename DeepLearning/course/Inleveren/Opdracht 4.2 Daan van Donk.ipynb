```python
mean_goals_per_minute = df.groupby('G')['Min'].mean()
mean_goals_per_minute
```

```python
std_goals_per_minute = df.groupby('G')['Min'].std()
std_goals_per_minute
```

### Merge

```python
df1 = mean_goals_per_minute.round(0).reset_index()
df2 = std_goals_per_minute.round(0).reset_index()
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2, on='G')
```

```python
df3
```

```python
df3.columns = ['Goals', 'Average Played Minutes', 'Played Minutes Standard Deviation']
```
